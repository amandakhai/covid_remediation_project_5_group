### Imports and Final CSV Reading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cdc_county = pd.read_csv("../data/complete_df.csv")
```

```python
cdc_county.head()
```

### Mobility EDA

```python
cdc_county["confirmed_cases_pct_change"] = cdc_county["daily_confirmed_cases"].pct_change()
cdc_county["deaths_pct_change"] = cdc_county["daily_deaths"].pct_change()
```

```python
mobility_columns = ["retail_and_recreation_percent_change_from_baseline",
                    "grocery_and_pharmacy_percent_change_from_baseline",
                    "parks_percent_change_from_baseline",
                    "transit_stations_percent_change_from_baseline",
                    "workplaces_percent_change_from_baseline",
                    "residential_percent_change_from_baseline"]
```

```python
mobility_df_avg = cdc_county[["date","County","confirmed_cases_pct_change","deaths_pct_change"]+mobility_columns].groupby("date").mean()
mobility_df_avg.fillna(0, inplace = True)
```

```python
plt.figure(figsize=(20,10))
color = ["red","blue","green","yellow","orange","purple"]
for i, columns in enumerate(mobility_columns):
    plt.plot(mobility_df_avg.index, mobility_df_avg[columns], color = color[i])
    plt.legend(mobility_columns)
plt.title(label = 'Average Mobility Timeseries', fontsize=20)
plt.axhline(0,color = "black",ls="--")
plt.xticks(ticks = ["2021-01-02","2021-02-01","2021-03-01","2021-04-01"],fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Period (Daily)", fontsize=18)
plt.ylabel("% Change", fontsize=18);
```

Looking at the graph above, we can see that aside from mobility within residential areas, most locations' % change in mobility vs baseline (baseline being pre-covid numbers) are negative. Average mobility for these areas have decreased by around -6.2%. However, with vaccine roll outs in 2021, an upward trend in mobility can be observed - especially for areas like parks, followed by retail and recreation, and groceries and pharmacies. This may be due to less social distancing regulations which allow for people to hangout in public, open-air spaces. Relative to other areas, transit stations' and workplace mobility is still below baseline numbers being that people are mostly still working from home. 

```python
mask = np.zeros_like(mobility_df_avg.corr())
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(10,10))
sns.heatmap(data = mobility_df_avg.corr(),
            annot=True, 
            vmin=-1, 
            vmax = 1,
            cmap="coolwarm",
            mask=mask)
plt.title("Mobility Correlation with Covid Deaths and Cases", size=12)
plt.tight_layout();
```

In the heatmap above, we can see a strong negative correlation between mobility in public areas and the numbers of covid cases and deaths. The intersections of our four mobility categories and two covid metrics show these relationships (highlighted in blue). What this means is that as cases rise, people are more likely to stay close to home. The reverse is also true - as the virus slows, weâ€™re seeing the public embrace a â€œnew normal,â€ returning to the public areas classified in our four categories.

### Regulation EDA

```python
flags = [col for col in cdc_county.columns if "flag" in col]
regulations = [" ".join(col.split("_")[1:]) for col in cdc_county.columns if any(i in col for i in ["c1", "c2", "c3", "c4", "c5", "c6", "c7", "e1", "h1", "h6", "h7", "h8"]) and col not in flags] 
```

```python
fig, ax = plt.subplots(nrows=12, ncols=1, figsize=(20,50))
ax = ax.ravel() # .ravel() turns a matrix into a vector, which is easier to iterate
for x, column in enumerate([0,1]):
    for i, dataframe in enumerate(flags_df):
        ax[i].plot(dataframe[dataframe.index.get_level_values(flags[i]) == column].index.get_level_values("date"),list(dataframe[dataframe.index.get_level_values(flags[i]) == column]["daily_confirmed_cases"]), color= ["red","purple"][x])
        ax[i].legend(["Regulation = False","Regulation = True"])
        ax[i].set_title(f'Average Confirmed Covid Cases based on {regulations[i]} Regulations')
        ax[i].set_xticks(ticks = ["2021-01-02","2021-02-01","2021-03-01","2021-04-01"])
        ax[i].set_ylabel("Average Cases")
        ax[i].set_xlabel('Period (Daily)')
        plt.tight_layout();
```

For both Covid deaths and confirmed cases, average numbers tend to be worse in counties where regulations regarding school closing, workplace closing, cancelling public events, restrictions on gatherings, stay at home requirements, movement restrictions, vaccination policy and protection of elderly people do not exist. 

When it comes to facial coverings however, we can see the opposite - this may be due to the fact that higher cases and deaths result in an increase in government intervention in trying to decrease the spread. The same trends can be observed in counties with income support and public transport regulations.
